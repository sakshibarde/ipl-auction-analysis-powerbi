```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/IPL Auction vs Player Performance Analysis using Python and Power BI/Merged_data.csv')
df.head()
```

```python
print("Rows:", df.shape[0])
print("Columns:", df.shape[1])
print("\nColumn names:")
print(df.columns.tolist())
```

```python
df.info()
df.isnull().sum().sort_values(ascending=False)
```

```python
df.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot for Auction Price
plt.figure(figsize=(8, 4))
sns.boxplot(data=df, x='Price_Cr', color='salmon')
plt.title("Boxplot of Auction Prices (Cr)")
plt.show()
```

```python
# Boxplot for ROI
sns.boxplot(data=df, x='ROI', color='skyblue')
plt.title("Boxplot of Return on Investment")
plt.show()
```

```python
from scipy.stats import zscore

# Calculate Z-score for Auction Price
df['Price_z'] = zscore(df['Price_Cr'])

# Filter values where Z > 3 or Z < -3
outliers = df[(df['Price_z'] > 3) | (df['Price_z'] < -3)]
print("Price Outliers:\n", outliers[['Name', 'Price_Cr', 'Team']])
```

```python
print("Roles:", df['Role_Stats'].unique())
print("Teams:", df['Team'].unique())
print("Countries:", df['National Side'].unique())
```

```python
# Strip whitespaces
df['Role_Stats'] = df['Role_Stats'].str.strip()

# Standardize similar roles
df['Role_Stats'] = df['Role_Stats'].replace({
    'Wicket-keeper': 'Wicket-Keeper'
})
print("Cleaned Roles:", df['Role_Stats'].unique())
df.to_csv('/content/drive/MyDrive/IPL Auction vs Player Performance Analysis using Python and Power BI/Merged_data.csv', index=False)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 4))
sns.countplot(data=df, x='Role_Stats', palette='Set2', order=df['Role_Stats'].value_counts().index)
plt.title('Distribution of Player Roles')
plt.ylabel('Number of Players')
plt.xlabel('Role')
plt.xticks(rotation=15)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='National Side', order=df['National Side'].value_counts().index, palette='viridis')
plt.title('Number of Players per Country')
plt.xlabel('Count')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.countplot(data=df, y='Team', order=df['Team'].value_counts().index, palette='pastel')
plt.title('Players in Each Team')
plt.xlabel('Number of Players')
plt.ylabel('Team')
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

top_exp = df.sort_values(by='Price_Cr', ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x='Name', y='Price_Cr', data=top_exp, palette='mako')
plt.title('Top 10 Most Expensive Players')
plt.xticks(rotation=45)
plt.ylabel('Price (in Cr)')
plt.show()
```

```python
top_roi = df[df['ROI'] > 0].sort_values(by='ROI', ascending=False).head(10)

plt.figure(figsize=(10, 5))
sns.barplot(x='Name', y='ROI', data=top_roi, palette='crest')
plt.title('Top 10 Best ROI Players')
plt.xticks(rotation=45)
plt.ylabel('ROI')
plt.show()
```

```python
sns.countplot(data=df, x='Role_Stats', palette='Set2')
plt.title('Distribution of Player Roles')
plt.xlabel('Role')
plt.ylabel('Number of Players')
plt.show()
```

```python
team_spend = df.groupby('Team')['Price_Cr'].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 5))
team_spend.plot(kind='bar', color='skyblue')
plt.title('Team-wise Total Auction Spend')
plt.ylabel('Total Spend (in Cr)')
plt.xticks(rotation=45)
plt.show()
```

```python
# Price vs Runs
sns.scatterplot(data=df, x='Price_Cr', y='RunsScored', hue='Role_Stats')
plt.title('Auction Price vs Runs Scored')
plt.show()

# Price vs Wickets
sns.scatterplot(data=df, x='Price_Cr', y='Wickets', hue='Role_Stats')
plt.title('Auction Price vs Wickets Taken')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(df[['Price_Cr', 'RunsScored', 'Wickets', 'ROI', 'Cost_per_Run', 'Cost_per_Wicket']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation between Auction Price & Performance')
plt.show()
```

```python
top_batsmen = df[df['Role_Stats'] == 'Batsman'].sort_values(by='RunsScored', ascending=False).head(5)
top_bowlers = df[df['Role_Stats'] == 'Bowler'].sort_values(by='Wickets', ascending=False).head(5)
top_allrounders = df[df['Role_Stats'] == 'All-Rounder'].sort_values(by='ROI', ascending=False).head(5)

print("Top Batsmen:\n", top_batsmen[['Name', 'RunsScored']])
print("Top Bowlers:\n", top_bowlers[['Name', 'Wickets']])
print("Top All-Rounders:\n", top_allrounders[['Name', 'ROI']])
```
