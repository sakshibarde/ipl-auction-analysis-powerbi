```python
import pandas as pd

# Load datasets
auction = pd.read_csv('/content/drive/MyDrive/IPL Auction vs Player Performance Analysis using Python and Power BI/ipl_2022_dataset.csv')
stats = pd.read_csv('/content/drive/MyDrive/IPL Auction vs Player Performance Analysis using Python and Power BI/IPL_Data.csv')
```

```python
# Preview
auction.head()
```

```python
stats.head()
```

```python
# Standardize player names in both
auction['Player'] = auction['Player'].str.lower().str.strip()
stats['Name'] = stats['Name'].str.lower().str.strip()
```

```python
# Rename for merging clarity
auction.rename(columns={
    'Player': 'Name',        # so both files have 'Name'
    'TYPE': 'Role_Auction',  # rename role column from auction
    'COST IN (CR.)': 'Price_Cr'
}, inplace=True)

stats.rename(columns={
    'Type': 'Role_Stats'     # rename role column from stats
}, inplace=True)

# Merge on 'Name'
merged = pd.merge(auction, stats, on='Name', how='inner')
```

```python
# Path to your project folder in Drive
path = '/content/drive/MyDrive/IPL Auction vs Player Performance Analysis using Python and Power BI/Merged_data.csv'

# Save merged CSV there
merged.to_csv(path, index=False)
merged.head()
```

```python
# Drop unnecessary/duplicate columns
merged.drop(columns=['Index', 'Team_y'], inplace=True)

# Rename Team_x to just Team
merged.rename(columns={'Team_x': 'Team'}, inplace=True)
```

```python
# Avoid divide-by-zero
merged['RunsScored'] = merged['RunsScored'].replace(0, 1)
merged['Wickets'] = merged['Wickets'].replace(0, 1)

# Cost KPIs
merged['Cost_per_Run'] = merged['Price_Cr'] / merged['RunsScored']
merged['Cost_per_Wicket'] = merged['Price_Cr'] / merged['Wickets']

# ROI formula: You can adjust weights here
merged['ROI'] = ((merged['RunsScored'] * 0.5) + (merged['Wickets'] * 25)) / merged['Price_Cr']

# Round
merged[['Cost_per_Run', 'Cost_per_Wicket', 'ROI']] = merged[['Cost_per_Run', 'Cost_per_Wicket', 'ROI']].round(2)

merged.to_csv('/content/drive/MyDrive/IPL Auction vs Player Performance Analysis using Python and Power BI/Merged_data.csv', index=False)
```

```python
# Check for null values in each column
merged.isnull().sum().sort_values(ascending=False)
```

```python
fill_zero_cols = [
    'RunsScored', 'Wickets', 'Overs', 'RunsConceded', 'InningsBowled',
    'Maidens', 'Mtc', '3s', '5s', 'CatchesTaken', 'StumpingsMade',
    'Ducks', '4s', '6s', '100s', '50s', 'NotOuts', 'InningsBatted',
    'MatchPlayed'
]

for col in fill_zero_cols:
    if col in merged.columns:
        merged[col] = merged[col].fillna(0)
```

```python
drop_cols = ['Full Name', 'Born', 'Age', 'Batting Style', 'Bowling', 'Url']
merged.drop(columns=drop_cols, inplace=True)
```

```python
# Fill BattingAVG & BattingS/R for batsmen/all-rounders
merged.loc[merged['Role_Stats'].isin(['Batsman', 'All-Rounder']), 'BattingAVG'] = merged.loc[merged['Role_Stats'].isin(['Batsman', 'All-Rounder']), 'BattingAVG'].fillna(merged['BattingAVG'].mean())
merged.loc[merged['Role_Stats'].isin(['Batsman', 'All-Rounder']), 'BattingS/R'] = merged.loc[merged['Role_Stats'].isin(['Batsman', 'All-Rounder']), 'BattingS/R'].fillna(merged['BattingS/R'].mean())

# Fill BowlingAVG & Economy for bowlers/all-rounders
merged.loc[merged['Role_Stats'].isin(['Bowler', 'All-Rounder']), 'BowlingAVG'] = merged.loc[merged['Role_Stats'].isin(['Bowler', 'All-Rounder']), 'BowlingAVG'].fillna(merged['BowlingAVG'].mean())
merged.loc[merged['Role_Stats'].isin(['Bowler', 'All-Rounder']), 'EconomyRate'] = merged.loc[merged['Role_Stats'].isin(['Bowler', 'All-Rounder']), 'EconomyRate'].fillna(merged['EconomyRate'].mean())
merged.loc[merged['Role_Stats'].isin(['Bowler', 'All-Rounder']), 'S/R'] = merged.loc[merged['Role_Stats'].isin(['Bowler', 'All-Rounder']), 'S/R'].fillna(merged['S/R'].mean())

merged.to_csv('/content/drive/MyDrive/IPL Auction vs Player Performance Analysis using Python and Power BI/Merged_data.csv', index=False)
```

```python
# Drop unnecessary columns
merged.drop(columns=['Sport', 'Best', 'R/O', 'HighestInnScore'], inplace=True)

# Fill 'National Side'
merged['National Side'] = merged['National Side'].fillna('Unknown')
```

```python
merged['Cost_per_Run'] = merged['Price_Cr'] / merged['RunsScored'].replace(0, 1)
merged['Cost_per_Wicket'] = merged['Price_Cr'] / merged['Wickets'].replace(0, 1)
merged['ROI'] = ((merged['RunsScored'] * 0.5) + (merged['Wickets'] * 25)) / merged['Price_Cr']

merged[['Cost_per_Run', 'Cost_per_Wicket', 'ROI']] = merged[['Cost_per_Run', 'Cost_per_Wicket', 'ROI']].round(2)
```

```python
merged[['BattingAVG', 'BattingS/R', 'BowlingAVG', 'EconomyRate', 'S/R']] = (
    merged[['BattingAVG', 'BattingS/R', 'BowlingAVG', 'EconomyRate', 'S/R']].fillna(0)
)
merged.to_csv('/content/drive/MyDrive/IPL Auction vs Player Performance Analysis using Python and Power BI/Merged_data.csv', index=False)
merged.isnull().sum().sort_values(ascending=False)
```
